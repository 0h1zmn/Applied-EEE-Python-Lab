```python
#Fundamental Logic Test 1

voltage = 3                                     #In Volts
current = 1.5                                   #In Amperes

resistance = voltage / current                  #Ohm's Law: R = V / I in Ohms
print("Resistance:", resistance, "Ohms")  
```

```python
#Fundamental Logic Test 2

resistors = [100, 220, 330, 50, 500, 680, 1000]        #Resistor values in Ohms

print("Resistor values (Ohms):", resistors)  

total_resistance = sum(resistors)                       #Total resistance in series is the sum of individual resistances
print("Total Resistance:", total_resistance, "Ohms")

print ("First resistor value:", resistors[0], "Ohms")   #Accessing the first resistor value using index 0

lowest_resistor = min(resistors)                        #Finding the lowest resistor value
print("Lowest Resistor Value:", lowest_resistor, "Ohms")

highest_resistor = max(resistors)                       #Finding the highest resistor value
print("Highest Resistor Value:", highest_resistor, "Ohms")

lowest_resistor_index = resistors.index(lowest_resistor)          #Finding the index of the lowest resistor value
print("Index of Lowest Resistor Value:", lowest_resistor_index)

highest_resistor_index = resistors.index(highest_resistor)       #Finding the index of the highest resistor value
print("Index of Highest Resistor Value:", highest_resistor_index)
```

```python
#Fundamental Logic Test 3

voltage = float(input("Enter the voltage of the battery: "))                #User input for voltage in volts
capacity = float(input("Enter the capacity of the battery in mAh: "))       #User input for capacity in milliampere-hours

energy = voltage * capacity / 1000                                          #Calculating energy stored in the battery in watt-hours (Wh)
print(f"The energy stored in the battery is {energy: .3f} Wh.")             #Output the energy stored in the battery formatted to 3 decimal places
```

```python
#Fundamental Logic Test 4

#Prompt for number of cells and validate: must be integer > 0
while True:
    #Try to convert input to integer and check if it's greater than 0
    try:
        #Convert input to integer
        cell_number = int(input("Enter the number of cells: "))
        if cell_number <= 0:  #Check if the number of cells is greater than 0
            print("Error: please enter an integer greater than 0.")
            continue    #If the input is valid,
        break           #exit the loop
    
    #Handle the case where input cannot be converted to an integer
    except ValueError:
        print("Error: please enter a valid integer greater than 0.")

#Initialize an empty list to store the voltages of the cells
voltages = []
#Loop to prompt for voltage of each cell and validate that it's a number
for i in range(cell_number):
    while True:
        try:
            #Prompt for voltage of each cell and convert to float
            voltage_input = input(f"Enter the voltage of cell {i + 1}: ")
            voltage = float(voltage_input)
            break

        except ValueError:
            print("Error: please enter a valid number for voltage.")
    
    #Add the valid voltage to the list of voltages
    voltages.append(voltage)

#Calculate the total voltage by summing the list of voltages
total_voltage = sum(voltages)
print(f"The total voltage of the cells is {total_voltage:.2f} V.")       #Output the total voltage formatted to 2 decimal places

#Calculate the average voltage by dividing the total voltage by the number of cells
average_voltage = sum(voltages) / len(voltages)
print(f"The average voltage of the cells is {average_voltage:.2f} V.")  #Output the average voltage formatted to 2 decimal places
```

```python
#Fundamental Logic Test 5

#Create a NumPy array of voltage readings from a battery pack
voltage_list = np.array([3.7, 3.6, 3.8, 2.9, 3.5, 3.4])
print("Engineer's voltage list:", voltage_list)

#Simulate charging the battery by adding 0.1 V to each cell
v_charge = voltage_list + 0.1
print("Voltage list after charging:", v_charge)

#Mean voltage, standard deviation, max, and min voltage calculations using NumPycalculation
mean_voltage = np.mean(voltage_list)
print(f"Mean voltage: {mean_voltage:.2f} V")

std_voltage = np.std(voltage_list)
print(f"Standard deviation of voltage: {std_voltage:.2f} V")

max_voltage = np.max(voltage_list)
print(f"Maximum voltage: {max_voltage:.2f} V")

min_voltage = np.min(voltage_list)
print(f"Minimum voltage: {min_voltage:.2f} V")

#Identify cells with voltage below 3.6 V and print their indices and values
bad_cells = voltage_list < 3.6

#print the bad cell values and their indices,
print("Bad cells (voltage < 3.6 V):", bad_cells)
print("Indices of bad cells:", np.where(bad_cells)[0] + 1)

#Identify critical cells with voltage below 3 V and print a warning if any are found
print("Critical cells (voltage < 3 V):", voltage_list[voltage_list < 3.0])
if np.min(voltage_list) < 3.0:
    print("Warning: Critical cell detected! Emergency Shutdown initiated.")
else:
    print("All cells are within safe voltage range.")
```
